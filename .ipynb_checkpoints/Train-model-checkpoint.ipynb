```python
import os
import numpy as np 
import cv2
import sys
import glob

import re

# https://stackoverflow.com/questions/4813061/non-alphanumeric-list-order-from-os-listdir
def sorted_aphanumeric(data):
    convert = lambda text: int(text) if text.isdigit() else text.lower()
    alphanum_key = lambda key: [ convert(c) for c in re.split('([0-9]+)', key) ] 
    return sorted(data, key=alphanum_key)



data = {}
data_dir = r'C:/Users/datng/Google Drive/pPlayground/cornhack2019/images/Fighting042_x264/'
# Image dimension
im_width = 180
im_height = 180
im_channel = 3
dim = im_width * im_height * im_channel
dtype = np.float32

all_files = [f for f in sorted_aphanumeric(glob.glob(data_dir+'/*.jpg'))]
# print(all_files)
n_files = len(all_files)

Y = np.zeros(n_files)
Y[290:1200] = 1

X = np.ndarray(shape=(n_files, im_width, im_height, im_channel), dtype=dtype)

for idx, file in enumerate(all_files):
    X[idx] = cv2.resize(cv2.imread(file), (im_width, im_height))


#shuffle dataset
idxs = np.random.permutation(n_files)
X = X[idxs]
Y = Y[idxs]

# # SPlit training and test set
data['X_train'] = X
# data['X_test'] = np.empty(shape=(num_val_test*n_cls,im_width, im_height, im_channel), dtype=datatype)

data['y_train'] = Y
# data['y_test'] = np.empty(num_val_test*n_cls).astype(int)
```

```python
class Dataset(object):
    def __init__(self, X, y, n_batchs):
        self.X = X
        self.y = y 
        self.n_batchs = n_batchs
        
    def to_dset(self):
        N = X.shape[0]
        idxs = np.random.permutation(N)
        self.X = self.X[idxs]
        self.y = self.y[idxs]
        
        item_per_batch = N // self.n_batchs
        batches = []
        
#         print('y',y.shape)
        for i in range(0,N, item_per_batch):
            if(i+item_per_batch > N): 
                batches.append((self.X[i:N],self.y[i:N]))
            else:
                batches.append((self.X[i:i+item_per_batch],self.y[i:i+item_per_batch]))
        return batches
        
```
